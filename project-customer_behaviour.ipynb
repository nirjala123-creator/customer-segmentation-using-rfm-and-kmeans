```python
rfm = df[["Customer ID", 
          "Days Since Last Purchase", 
          "Items Purchased", 
          "Total Spend"]].copy()

rfm.columns = ["CustomerID", "Recency", "Frequency", "Monetary"]

rfm.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
rrm_scaled= scaler.fit_transform(rfm[["Recency", "Frequency", "Monetary"]])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
rfm["Cluster"] = kmeans.fit_predict(rrm_scaled)
```

```python
rfm.groupby("Cluster")[["Recency","Frequency","Monetary"]].mean()
```

Cluster 2
Lowest Recency (11 days)
Highest Frequency (20 purchases)
Highest Monetary (1459 spend)Cluster 2 represents High-Value Loyal Customers. 
These customers contribute maximum revenue and show strong engagement with the business.

Cluster 1 consists of Regular Customers who generate steady revenue and
maintain consistent purchasing behavior.

Cluster 0 represents At-Risk or 
Low-Value Customers who show lower engagement and may indicate potential churn risk.
